```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
data = data = pd.read_csv('C:\\Users\\solomon kure\\Desktop\\data.csv', index_col = 0)
data.drop(columns = ['Unnamed: 32'], inplace = True)
data.head()
# Splitting the data
X = data.drop(columns = ['diagnosis'])
y = data['diagnosis']
# Training the model
knn = KNeighborsClassifier()
knn.fit(X,y)
# Finding our predictions
prediction = knn.predict(X)
prediction
# Checking accuracy of our model accuracy
knn = KNeighborsClassifier(n_neighbors = 10)
X_train,X_test,y_train,y_test = train_test_split(X , y , test_size = .3,random_state = 21,stratify = y)
knn.fit(X_train,y_train)
y_prediction = knn.predict(X_test)
knn.score(X_test,y_test)
```
